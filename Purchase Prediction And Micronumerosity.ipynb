# **ENHANCEMENT SUGGESTED BY CHATGPT**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load data
url = 'https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Customer%20Purchase.csv'
purchase = pd.read_csv(url)

# Explore data
purchase.head()
purchase.info()
purchase.describe()
purchase.shape
purchase.columns

# Prepare data
x = purchase[['Age', 'Gender', 'Education', 'Review']]
y = purchase['Purchased']

# Label encoding
le = LabelEncoder()
x['Gender'] = le.fit_transform(x['Gender'])
x['Education'] = le.fit_transform(x['Education'])
x['Review'] = le.fit_transform(x['Review'])

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=2529)

# Model training
model = RandomForestClassifier()
model.fit(x_train, y_train)

# Predictions
y_pred = model.predict(x_test)

# Model evaluation
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_rep)
```

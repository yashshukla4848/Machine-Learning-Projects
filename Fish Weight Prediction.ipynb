```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# Load data
url = "https://github.com/YBIFoundation/Dataset/raw/main/Fish.csv"
fish = pd.read_csv(url)

# Explore data
fish.head()
fish.info()
fish.describe()
fish.columns
fish.shape

# Visualize relationships between features and target
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, ax in enumerate(axes.flat):
    if i < len(fish.columns) - 1:
        ax.scatter(fish[fish.columns[i]], fish['Weight'])
        ax.set_title(f'{fish.columns[i]} vs Weight')
        ax.set_xlabel(fish.columns[i])
        ax.set_ylabel('Weight')

plt.tight_layout()
plt.show()

# Handle categorical feature (Category)
encoder = OneHotEncoder(drop='first', sparse=False)
category_encoded = pd.DataFrame(encoder.fit_transform(fish[['Category']]), columns=encoder.get_feature_names_out(['Category']))
fish = pd.concat([fish, category_encoded], axis=1)
fish.drop(['Category'], axis=1, inplace=True)

# Check the updated column names
print(fish.columns)

# Prepare data
x = fish[['Height', 'Width', 'Length1', 'Length2', 'Length3'] + list(category_encoded.columns)]
y = fish['Weight']

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=2529)

# Model training
model = LinearRegression()
model.fit(x_train, y_train)

# Predictions
y_pred = model.predict(x_test)

# Model evaluation
print("mean_absolute_percentage_error =", mean_absolute_percentage_error(y_test, y_pred))
print("mean_absolute_error =", mean_absolute_error(y_test, y_pred))
print("mean_squared_error =", mean_squared_error(y_test, y_pred))
```
